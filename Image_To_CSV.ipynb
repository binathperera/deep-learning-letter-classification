```python
from PIL import ImageTk, Image
import numpy as np
import sys
import os
import csv
```

```python
def createFileList(myDir, format='.png'):
    fileList = []
    print(myDir)
    labels = []
    names = []
    keywords = {"F" : "1","S": "0",} # keys and values to be changed as needed
    for root, dirs, files in os.walk(myDir, topdown=True):
        for name in files:
            if name.endswith(format):
                fullName = os.path.join(root, name)
                fileList.append(fullName)
            for keyword in keywords:
                if keyword in name:
                    labels.append(keywords[keyword])
                else:
                    continue
            names.append(name)
    return fileList, labels, names
# load the original image
myFileList, labels, names  = createFileList('Content/')
i = 0
for file in myFileList:
    print(file)
    img_file = Image.open(file).convert('RGB')
    # img_file.show()
# get original image parameters...
    arr = np.array(img_file)
    # record the original shape
    shape = arr.shape
    # make a 1-dimensional view of arr
    value = arr.ravel()
    # convert it to a matrix

# Make image Greyscale
    #img_grey.save('result.png')
    #img_grey.show()
# Save Greyscale values
    i +=1
    
    print(value)
    f = open("letters.csv", 'a', newline='')
    writer = csv.writer(f)
    writer.writerow(value)
```
